```python
import torch
import torch.nn as nn
import torchaudio
import numpy as np

print("Imports done!")
```

```python
# batch of grayscale spectrograms
tensor = torch.randn(32, 1, 64, 64)

# flatten everything except batch dimension
flattened = tensor.view(32, -1)

print("Original shape :", tensor.shape)
print("Flattened shape:", flattened.shape)
print("Input features for Linear layer:", flattened.shape[1])
```

```python
# 1 second audio at 16000 Hz
sample_rate = 16000
audio       = torch.randn(1, sample_rate)

# mel spectrogram config
mel_transform = torchaudio.transforms.MelSpectrogram(
    sample_rate = sample_rate,
    n_fft       = 400,
    hop_length  = 160,
    n_mels      = 64
)

mel = mel_transform(audio)

print("Audio shape      :", audio.shape)
print("Mel Spectrogram  :", mel.shape)
print("Answer → (1, 64, 101)")
```

```python
# input shape (1, 64, 64)
x = torch.randn(1, 1, 64, 64)

# conv layer with padding=1 keeps same size
conv = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1)

# maxpool halves the size
pool = nn.MaxPool2d(kernel_size=2, stride=2)

out = pool(conv(x))

print("Input shape  :", x.shape)
print("After Conv   :", conv(x).shape)
print("After MaxPool:", out.shape)
print("Spatial output (H, W):", out.shape[2], out.shape[3])
print("Answer → (32, 32)")
```

```python
# linear layer
layer = nn.Linear(in_features=128, out_features=64)

weights = layer.weight.numel()
biases  = layer.bias.numel()
total   = weights + biases

print("Weights    :", weights)
print("Biases     :", biases)
print("Total params:", total)
print("Answer → 8,256")
```

```python
import math

total_files = 1050
batch_size  = 32
drop_last   = False

num_batches = math.ceil(total_files / batch_size)

print("Total files :", total_files)
print("Batch size  :", batch_size)
print("Num batches :", num_batches)
print("Answer → 33")
```

```python
# model output logits and target class
logits = torch.tensor([[2.5, 1.0, 0.1]])
target = torch.tensor([0])

criterion = nn.CrossEntropyLoss()
loss      = criterion(logits, target)

print("Logits :", logits)
print("Target : Class 0")
print("Loss   :", round(loss.item(), 3))
print("Answer → 0.273")
```

```python
# SGD update formula: new_weight = weight - lr * gradient
current_weight = 0.5
gradient       = 0.2
learning_rate  = 0.01

new_weight = current_weight - learning_rate * gradient

print("Current weight :", current_weight)
print("Gradient       :", gradient)
print("Learning rate  :", learning_rate)
print("New weight     :", new_weight)
print("Answer → 0.498")
```

```python
# formula: padding = (kernel_size - 1) / 2
kernel_size = 5
padding     = (kernel_size - 1) // 2

# verify with actual conv
x    = torch.randn(1, 1, 64, 64)
conv = nn.Conv2d(1, 1, kernel_size=5, stride=1, padding=padding)
out  = conv(x)

print("Kernel size    :", kernel_size)
print("Padding needed :", padding)
print("Input shape    :", x.shape)
print("Output shape   :", out.shape)
print("Answer → 2")
```
